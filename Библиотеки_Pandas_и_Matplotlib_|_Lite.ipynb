### Задание 1

Ниже дано 2 графика. Сделайте их более простыми для восприятия:
- подпишите оси - "ось X" и "X и Y"
- подпишите название графика - "sin(x)/2 и cos(x)"
- увеличьте размер полотна

````python
# Исходные графики:

import matplotlib.pyplot as plt    # Импорт модуля pyplot
import numpy as np                 # Импорт библиотеки numpy

x = np.linspace(0, 10, 101)        # Генерация последовательности из 101 числа с шагом в 0.1
y = np.cos(x)                      # Запись cos(x) в переменную y
z = np.sin(x) / 2                  # Запись sin(x) / 2 в переменную z

plt.plot(x, y)                     # Построение графика x, y
plt.plot(x, z)                     # Построение графика x, z
plt.show()                         # Фиксация графика
````

````python
# Ваше решение

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 101)
y = np.cos(x)
z = np.sin(x) / 2

plt.figure(figsize=(12,4))
plt.plot(x, y, 'c:', label='cos(x)', linewidth=3, color='lightgreen',
         marker='+', markeredgecolor='black', markersize=5, markevery=20)
plt.plot(x, z, label='sin(x)/2', linewidth=2, color='orange',
         marker='1', markeredgecolor='black', markersize=5, markevery=10)
plt.title('cos(x), sin(x)/2')
plt.legend(loc='lower center')
plt.ylabel('Ось Y')
plt.xlabel('Ось X')
plt.show()    
````
