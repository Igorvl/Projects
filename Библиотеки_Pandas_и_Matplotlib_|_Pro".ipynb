Дан следующий исходный график:

```python
import matplotlib.pyplot as plt           # Импорт модуля pyplot
import numpy as np                        # Импорт библиотеки numpy
x = np.linspace(0, 20, 101)               # Генерация последовательности из 101 числа с шагом в 0.2
y = np.cos(x)                             # Запись cos(x) в переменную y

plt.plot(x, y)                            # Построение графика x, y
plt.show()                                # Фиксация графика
```

- размер полотна: 12,4
- ширина линии: 5
- цвет линии: **4b1234**
- тип маркера: h
- размер маркера: 10
- цвет маркера: **ff00d5**
- периодичность маркера: 5
- легенда графика закреплена в правом нижнем углу: **cos(x)**
- заголовок графика: **cos(x)**
- подписи осей: **ось X** и **ось Y**
- отрисована сетка

```python
# Ваше решение
import matplotlib.pyplot as plt # Импорт модуля pyplot
import numpy as np # Импорт библиотеки numpy
x = np.linspace(0, 20, 101) # Генерация последовательности из 101 числа с шагом в 0.2
y = np.cos(x) # Запись cos(x) в переменную y

plt.figure(figsize=(12, 4)) # Изменение размера полотна графика
plt.plot(x, y, color='#4b1234', linewidth=5, marker='h', markersize=10, markerfacecolor='#ff00d5', markevery=5, label='cos(x)')
plt.legend(loc='lower right') # Легенда графика закреплена в правом нижнем углу
plt.title('cos(x)') # Заголовок графика
plt.xlabel('ось X') # Подпись оси X
plt.ylabel('ось Y') # Подпись оси Y
plt.grid() # Отрисована сетка
plt.show() # Фиксация графика
```

Создайте резервную копию датафрейма с помощью специального метода. Удалите из датафрейма все строки, где есть хотя бы один пропуск.

Затем сохраните в переменную все уникальные категории приложений (столбец **Category**). Используя цикл, вычислите количество приложений для каждой категории.

```python
# Ваше решение

# Создание резервной копии датафрейма
df_backup = df.copy()

# Удаление строк с пропусками
df.dropna(inplace=True)

# Получение уникальных категорий приложений
unique_categories = df["Category"].unique()

# Вычисление количества приложений для каждой категории
category_counts = {}
for category in unique_categories:
    category_count = df[df["Category"] == category].shape[0]
    category_counts[category] = category_count

display(category_counts)
```

Выведите среднее значение и медианное значение рейтинга для каждой категории. Для этого вам понадобится выполнить сортировку по столбцу **Category**, используя метод `.groupby()` и применить метод `.agg()`.

```python
# Ваше решение
grouped = df.groupby(by='Category')
aggregated = grouped.agg({'Rating': ['mean', 'median']})
aggregated.reset_index(inplace=True)
print(aggregated)
```
